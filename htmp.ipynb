```python
import seaborn as sns
```

```python
iris = sns.load_dataset('iris')
species = iris.pop('species')
sns.clustermap(iris)
```

```python
#cambiando el tamao de la figura de calor

sns.clustermap(
    iris,
    figsize=(7,5),
    row_cluster=False,
    dendrogram_ratio=(.1, .2),
    cbar_pos=(0, .2, .03, .4)
)
```

```python
# aadiendo colores a las itequetas, para identificar observaciones y detalles

lut = dict(zip(species.unique(), 'rgb'))
row_colors = species.map(lut)
sns.clustermap(iris, row_colors=row_colors)
```

```python
# usando una paleta de color diferentes, para variar "va much"

sns.clustermap(iris, cmap='mako', vmin=0, vmax=10)
```

```python
# utilizando distintos parametros de clustering

sns.clustermap(iris, metric='correlation', method='single')
```

```python
# estandarizando la data dentro de las columnas

sns.clustermap(iris, standard_scale=1)
```

```python
# normalizando la data dentro de los registros

sns.clustermap(iris, z_score=0, cmap='vlag', center=0)
```
